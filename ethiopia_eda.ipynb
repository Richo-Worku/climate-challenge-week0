```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set_style("whitegrid")
```

```python
df = pd.read_csv("data/ethiopia.csv")
df.head()
```

```python
df = df.replace(-999, np.nan)
```

```python
df["Country"] = "Ethiopia"
```

```python
df["DATE"] = pd.to_datetime(df["YEAR"] * 1000 + df["DOY"], format="%Y%j")
df.head()
```

```python
df["Month"] = df["DATE"].dt.month
df.head()
```

```python
df.duplicated().sum()
```

## Summary of  Statistics Interpretation

The dataset shows that Ethiopia experiences moderate temperature conditions, with an average temperature of approximately 16°C. Temperatures range from about 10°C to 21°C, indicating noticeable seasonal variation.

Rainfall is highly variable, with most days receiving little to no precipitation, but occasional extreme rainfall events reaching up to 82 mm. This suggests irregular but significant rainy periods.

Humidity levels are generally moderate (around 68%) but fluctuate widely depending on the season. Wind speed remains relatively low and stable throughout the dataset.

Overall, the climate data reflects a typical seasonal pattern with moderate temperatures, distinct wet and dry periods, and occasional extreme weather events.

```python
missing = df.isna().sum()
missing_pct = (missing / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing": missing,
    "Percent": missing_pct
}).sort_values("Percent", ascending=False)

missing_report
```

No missing values per column with  with >5% nulls.


```python
from scipy.stats import zscore

cols = ["T2M", "T2M_MAX", "T2M_MIN", "PRECTOTCORR", "RH2M", "WS2M", "WS2M_MAX"]

z_scores = df[cols].apply(zscore)
outliers = (z_scores.abs() > 3)
outlier_count = outliers.any(axis=1).sum()
print("Outlier rows:", outlier_count)
df_outliers = df[outliers.any(axis=1)]
df_outliers.head()
```

## Outlier Detection

Z-scores were computed for key climate variables (T2M, T2M_MAX, T2M_MIN, PRECTOTCORR, RH2M, WS2M, WS2M_MAX) to identify extreme values.

Rows with absolute Z-scores greater than 3 were flagged as outliers.

A total of 132 rows were identified as containing outliers.

These outliers may represent extreme weather events such as heatwaves or heavy rainfall, which are important for climate analysis. Therefore, removing them is not reasonable. 

```python
monthly_temp = df.groupby("Month")["T2M"].mean()
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.plot(monthly_temp.index, monthly_temp.values, marker="o")

plt.title("Monthly Average Temperature (2015–2026)")
plt.xlabel("Month")
plt.ylabel("Temperature (°C)")
```

```python
# warmest & coolest months
warmest_month = monthly_temp.idxmax()
coolest_month = monthly_temp.idxmin()

warmest_value = monthly_temp.max()
coolest_value = monthly_temp.min()
# Highlight points (this is key)
plt.scatter(warmest_month, warmest_value)
plt.scatter(coolest_month, coolest_value)

# Annotate warmest
plt.annotate(f"Warmest ({warmest_month})",
             xy=(warmest_month, warmest_value),
             xytext=(warmest_month, warmest_value + 0.5),
             arrowprops=dict(),
             ha='center')

# Annotate coolest
plt.annotate(f"Coolest ({coolest_month})",
             xy=(coolest_month, coolest_value),
             xytext=(coolest_month, coolest_value - 0.5),
             arrowprops=dict(),
             ha='center')
```

```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
plt.bar(monthly_rain.index, monthly_rain.values)

plt.title("Monthly Total Rainfall (2015–2026)")
plt.xlabel("Month")
plt.ylabel("Precipitation")
peak_months = monthly_rain.sort_values(ascending=False).head(3)

print(peak_months)
```
